```python
#1_ Write a Python program to generate a random variable and display its value

import numpy as np

# Generate a single random float between 0 and 1
random_value = np.random.rand()
print("Random Variable:", random_value)
```

```python
#2_  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters
low, high = 1, 7  # Discrete uniform from 1 to 6 (like dice)

x = np.arange(low, high)
pmf_vals = randint.pmf(x, low, high)

plt.stem(x, pmf_vals, use_line_collection=True)
plt.title("PMF of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("PMF")
plt.grid()
plt.show()
```

```python
#3_  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

from scipy.stats import bernoulli

def bernoulli_pdf(p):
    x = [0, 1]
    return {val: bernoulli.pmf(val, p) for val in x}

# Example
print("Bernoulli PDF (p=0.3):", bernoulli_pdf(0.3))
```

```python
#4_Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.

from scipy.stats import binom

data = binom.rvs(n=10, p=0.5, size=1000)

plt.hist(data, bins=np.arange(12) - 0.5, density=True, edgecolor='black')
plt.title("Binomial Distribution Histogram (n=10, p=0.5)")
plt.xlabel("Successes")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
#5_Create a Poisson distribution and visualize it using Python.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

mu = 4
x = np.arange(0, 15)
pmf_vals = poisson.pmf(x, mu)

# Plot without 'use_line_collection'
plt.stem(x, pmf_vals)
plt.title("PMF of Poisson Distribution (=4)")
plt.xlabel("k")
plt.ylabel("P(X=k)")
plt.grid()
plt.show()
```

```python
#6_ Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

cdf_vals = randint.cdf(x, low, high)

plt.step(x, cdf_vals, where='post')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("CDF")
plt.grid()
plt.show()
```

```python
#8_ Simulate data from a normal distribution and plot its histogram.

from scipy.stats import norm

data = norm.rvs(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, density=True, edgecolor='black')
plt.title("Normal Distribution (mean=0, std=1)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid()
plt.show()
```

```python
#9_ Write a Python function to calculate Z-scores from a dataset and plot them

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = [(x - mean) / std for x in data]
    return z_scores

data = np.random.normal(loc=50, scale=10, size=1000)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30, edgecolor='black')
plt.title("Z-Scores Histogram")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
#10_ Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

sample_means = []

for _ in range(1000):
    sample = np.random.exponential(scale=2.0, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, edgecolor='black', density=True)
plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid()
plt.show()
```

```python
#11_Simulate multiple samples from a normal distribution and verify the Central Limit Theorem

import numpy as np
import matplotlib.pyplot as plt

sample_means = []

for _ in range(1000):
    sample = np.random.normal(loc=50, scale=15, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, edgecolor='black', density=True)
plt.title("CLT: Sampling Distribution of the Mean")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid()
plt.show()
```

```python
#13_ Generate random variables and calculate their corresponding probabilities using the binomial distribution

from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, 11)
probs = binom.pmf(x, n, p)

for xi, pi in zip(x, probs):
    print(f"P(X={xi}) = {pi:.4f}")
```

```python
#14_ Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

def calculate_z_score(x, mean, std):
    return (x - mean) / std

x = 75
mean = 70
std = 10
z = calculate_z_score(x, mean, std)
print("Z-score:", z)
print("P-value (from standard normal):", 1 - norm.cdf(z))
```

```python
#15_ Implement hypothesis testing using Z-statistics for a sample dataset

# H0: mu = 100, sample mean = 105, sigma = 15, n = 36
sample_mean = 105
mu = 100
sigma = 15
n = 36

z_stat = (sample_mean - mu) / (sigma / np.sqrt(n))
p_val = 1 - norm.cdf(z_stat)

print("Z-statistic:", z_stat)
print("P-value:", p_val)
```

```python
#16_ Create a confidence interval for a dataset using Python and interpret the result

import scipy.stats as stats

data = np.random.normal(50, 10, 100)
mean = np.mean(data)
std_err = stats.sem(data)
conf_int = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)

print("95% Confidence Interval for the Mean:", conf_int)
```

```python
#17_ Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

data = np.random.normal(80, 5, 50)
mean = np.mean(data)
ci = stats.norm.interval(0.95, loc=mean, scale=np.std(data)/np.sqrt(len(data)))
print("95% CI for Normal Data Mean:", ci)
```

```python
#21_ Write a Python program to compare the standard deviations of two datasets and visualize the difference

data1 = np.random.normal(60, 10, 100)
data2 = np.random.normal(60, 20, 100)

plt.hist(data1, bins=30, alpha=0.5, label='Std Dev = 10')
plt.hist(data2, bins=30, alpha=0.5, label='Std Dev = 20')
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.grid()
plt.show()
```

```python
#22_ Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

data = np.random.normal(100, 15, 1000)

data_range = np.max(data) - np.min(data)
iqr = np.percentile(data, 75) - np.percentile(data, 25)

print("Range:", data_range)
print("Interquartile Range (IQR):", iqr)
```

```python
#23_ Implement Z-score normalization on a dataset and visualize its transformation

data = np.random.normal(30, 5, 1000)
z_scores = (data - np.mean(data)) / np.std(data)

plt.hist(z_scores, bins=30, edgecolor='black')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```
